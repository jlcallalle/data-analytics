#1. Cambio del tipo de dato de una columna

```python
import re
```

```python
def eliminar_coma(texto):

  texto_limpio = re.sub(r',', '', str(texto))

  return texto_limpio
```

```python
dataset_arbitrios_jm["MONTO_SERENAZGO"] = dataset_arbitrios_jm["MONTO_SERENAZGO"].apply(eliminar_coma)
```

```python
dataset_arbitrios_jm.info()
```

```python
dataset_arbitrios_jm["DESCUENTO_TOPE_SERENAZGO"] = dataset_arbitrios_jm["DESCUENTO_TOPE_SERENAZGO"].apply(eliminar_coma)
```

```python
dataset_arbitrios_jm.head()
```

```python
dataset_arbitrios_jm["MONTO_SERENAZGO"] = dataset_arbitrios_jm["MONTO_SERENAZGO"].astype(float)
```

```python
dataset_arbitrios_jm["DESCUENTO_TOPE_SERENAZGO"] = dataset_arbitrios_jm["DESCUENTO_TOPE_SERENAZGO"].astype(float)
```

```python
dataset_arbitrios_jm.info()
```

#7. Agrupamiento

```python
dataset_arbitrios_jm.groupby('PERIODO').MONTO_SERENAZGO.sum()
```

```python
dataset_arbitrios_jm.groupby('PERIODO').agg({'MONTO_SERENAZGO': 'sum', 'MONTO_PARQUE_JARDIN': 'mean'})
```

```python
dataset_arbitrios_jm.groupby('PERIODO').agg(MTO_TOTAL_SERENAZGO=('MONTO_SERENAZGO', 'sum'),
                                            MTO_PROMEDIO_PARQUE_JARDIN=('MONTO_PARQUE_JARDIN', 'mean'))
```

```python
dataset_arbitrios_jm.groupby(by=['PERIODO'], as_index=False).agg(MTO_TOTAL_SERENAZGO=('MONTO_SERENAZGO', 'sum'),
                                            MTO_PROMEDIO_PARQUE_JARDIN=('MONTO_PARQUE_JARDIN', 'mean'))
```
